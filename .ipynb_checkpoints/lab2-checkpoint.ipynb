```python
alph = ['а', 'б', 'в', 'г', 'д', 'е', 'ж', 'з', 'и', 'й', 'к', 'л', 'м', 'н', 'о', 'п', 'р', 'с', 'т', 'у', 'ф', 'х', 'ц', 'ч', 'ш', 'щ', 'ы', 'ь', 'э', 'ю', 'я']
alph_len = len(alph)
```

```python
def format_text(text):
    text_f = ''.join(re.findall(u"[а-яА-Я]", text)).lower()
    text_f = text_f.replace('ё', 'е')
    text_f = text_f.replace('ъ', 'ь')
    return text_f
```

```python
def monogram_freq(text):
    freq = np.zeros(alph_len)
    for i in text:
        freq[alph.index(i)] += 1
    return freq/freq.sum()
```

```python
a = monogram_freq(text)
print(a)
len(a)
```

```python
def bigram_freq(text):
    freq = np.zeros((alph_len, alph_len))
    length = len(text) - 1
    for i in range(length):
        freq[alph.index(text[i])][alph.index(text[i + 1])] += 1
    return freq/freq.sum().sum()
```

```python
bigram_freq(text)
```

```python
def entropy(text, l = 1):
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    ent = 0
    if l == 1:
        freq = monogram_freq(text)
        for i in range(alph_len):
            ent += -freq[i] * math.log(freq[i], 2)
    elif l == 2:
        freq = bigram_freq(text)
        for i in range(alph_len):
            for j in range(alph_len):
                if freq[i][j] != 0:
                    ent += -freq[i][j] * math.log(freq[i][j], 2)
    ent = ent / l
    return ent
```

```python
def index(text, l = 1):
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    index = 0
    text_len = len(text)
    if l == 1:
        freq = monogram_freq(text)
        freq = freq * text_len
        for i in range(alph_len):
            index += freq[i] * (freq[i] - 1)
    elif l == 2:
        freq = bigram_freq(text)
        freq = freq * (text_len - 1)
        for i in range(alph_len):
            for j in range(alph_len):
                 index += freq[i][j] * (freq[i][j] - 1)
    index = index / (text_len * (text_len - 1))
    return index
```

```python
def cut_into_texts(text, L = 10, N = 10000):
    text_len = len(text)
    if L * N > text_len:
        raise ValueError('Text don"t have enough symbols.')
    point = 0
    texts = ['']*N
    for i in range(N):
        texts[i] = text[point : (point + L)]
        point += L
    return texts
```

```python
def text_corrupt_vigenere(text, r = 5):
    new_text = ''
    key = np.random.randint(low = 0, high = alph_len - 1, size = r)
    for i in text:
        y = (alph.index(i) + key[alph.index(i) % r]) % alph_len
        new_text += alph[y]
    return new_text
```

```python
def text_corrupt_rand(text, l = 1):
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    text_len = len(text)
    new_text = ''
    if (l == 1):
        for i in text:
            y = random.randint(0, alph_len - 1)
            new_text += alph[y]
    elif (l == 2):
        i = 0
        while i <= text_len:
            y1 = random.randint(0, alph_len - 1)
            y2 = random.randint(0, alph_len - 1)
            new_text += alph[y1]
            new_text += alph[y2]
            i = i + 2
    return new_text
```

```python
def text_corrupt_affine(text, l = 1):
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    text_len = len(text)
    new_text = ''
    if (l == 1):
        a = random.randint(0, alph_len - 1)
        b = random.randint(0, alph_len - 1)
        for i in text:
            y = (a * alph.index(i) + b) % alph_len
            new_text += alph[y]
    elif (l == 2):
        i = 0
        a = random.randint(0, (alph_len - 1)^2)
        b = random.randint(0, (alph_len - 1)^2)
        while i < text_len - 1:
            temp = (a * (alph.index(text[i])*alph_len + alph.index(text[i + 1])) + b) % ((alph_len)^2)
            y1 = temp // alph_len
            y2 = temp % alph_len
            new_text += alph[y1]
            new_text += alph[y2]
            i = i + 2
    return new_text
```

```python
def text_corrupt_reccur(text, l = 1):
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    text_len = len(text)
    new_text = ''
    if (l == 1):
        new_text += alph[random.randint(0, alph_len - 1)]
        new_text += alph[random.randint(0, alph_len - 1)]
        for i in range(2, text_len):
            y = (alph.index(new_text[i - 1]) + alph.index(new_text[i - 2])) % alph_len
            new_text += alph[y]
    elif (l == 2):
        i = 0
        in1 = random.randint(0, (alph_len - 1)^2)
        in2 = random.randint(0, (alph_len - 1)^2)
        new_text = new_text + alph[in1 // alph_len] + alph[in1 % alph_len]
        new_text = new_text + alph[in2 // alph_len] + alph[in2 % alph_len]
        i = 4
        while i < text_len - 1:
            prev1 = alph.index(new_text[i - 4])*alph_len + alph.index(new_text[i - 3])
            prev2 = alph.index(new_text[i - 2])*alph_len + alph.index(new_text[i - 1])
            temp = (prev1 + prev2) % alph_len^2
            y1 = temp // alph_len
            y2 = temp % alph_len
            new_text += alph[y1]
            new_text += alph[y2]
            i = i + 2
    return new_text
```

```python
def top_freq(text, h = 1, l = 1):
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    top = ['']*h
    if l == 1:
        freq = monogram_freq(text)
        for i in range(h):
            pos = np.argmax(freq)
            freq[pos] = 0
            top[i] = alph[pos]
    elif l == 2:
        freq = bigram_freq(text)
        for i in range(h):
            pos = np.argmax(freq)
            pos1 = pos // alph_len
            pos2 = pos % alph_len
            freq[pos1][pos2] = 0
            top[i] = alph[pos1] + alph[pos2]
    return top
```

```python
def get_texts(text, f = text_corrupt_vigenere, L = 100, N = 1000):
    good_texts = cut_into_texts(text, L, N)
    bad_texts = []
    for i in good_texts:
        bad_texts.append(f(i))
    return good_texts, bad_texts
```

```python
def criterion_2_0(test, top, l = 1):
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    state = 1
    for i in top:
        if i not in test:
            state = 0
            return state
    return state
```

```python
def forall_2_0(text, good_texts, bad_texts, l = 1, h = 3):
    false_pos = 0
    false_neg = 0
    top = top_freq(text, h, l)
    N = len(good_texts)
    for i in range(N):
        if (criterion_2_0(good_texts[i], top, l) == 0):
            false_neg += 1
        if (criterion_2_0(bad_texts[i], top, l) == 1):
            false_pos += 1
    return false_neg/N, false_pos/N
```

```python
good_texts, bad_texts = get_texts(text, text_corrupt_vigenere)
forall_2_0(text, good_texts, bad_texts, h = 3)
```

```python
def criterion_2_1(test, top, l = 1, k = 5, h = 7):
    af = []
    status = 1
    if (h <= k):
         raise ValueError('h must be > k.')
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    for i in top:
        if i in test:
            af.append(i)
    intersect = 0
    for i in top:
        if i in af:
            intersect += 1
    if intersect < k:
        status = 0
        return status
    else:
        return status
```

```python
def forall_2_1(text, good_texts, bad_texts, l = 1, k = 5, h = 7 ):
    false_pos = 0
    false_neg = 0
    N = len(good_texts)
    top = top_freq(text, h, l)
    for i in range(N):
        if (criterion_2_1(good_texts[i], top, l, k, h) == 0):
            false_neg += 1
        if (criterion_2_1(bad_texts[i], top, l, k, h) == 1):
            false_pos += 1
    return false_neg/N, false_pos/N
```

```python
def criterion_2_2(test, top, freq, l = 1):
    status = 1
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    if (l == 1):
        emp_freq = monogram_freq(test)
        for i in top:
            if (emp_freq[alph.index(i)] < freq[alph.index(i)]):
                status = 0
                return status
    elif (l == 2):
        emp_freq = bigram_freq(test)
        for i in top:
            i1 = i[0]
            i2 = i[1]
            if (emp_freq[alph.index(i1)][alph.index(i2)] < freq[alph.index(i1)][alph.index(i2)]):
                status = 0
                return status
    return status
```

```python
def forall_2_2(text, good_texts, bad_texts, l = 1, h = 3):
    false_pos = 0
    false_neg = 0
    top = top_freq(text, h, l)
    N = len(good_texts)
    if (l == 1):
        freq = monogram_freq(text)
    elif (l == 2):
        freq = bigram_freq(text)
    for i in range(N):
        if (criterion_2_2(good_texts[i], top, freq, l) == 0):
            false_neg += 1
        if (criterion_2_2(bad_texts[i], top, freq, l) == 1):
            false_pos += 1
    return false_neg/N, false_pos/N
```

```python
def criterion_2_3(test, top, freq, l = 1, h = 3):
    status = 1
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    F = 0
    for i in top:
        if i in test:
            F += 1
    K = 0
    if (l == 1):
        for i in top:
            K += freq[alph.index(i)]
    elif (l == 2):
        for i in top:
            i1 = i[0]
            i2 = i[1]
            K += freq[alph.index(i1)][alph.index(i2)]
    if (F < K):
        status = 0
        return status
    else:
        return status
```

```python
def forall_2_3(text, good_texts, bad_texts, l = 1, h = 3):
    false_pos = 0
    false_neg = 0
    top = top_freq(text, h, l)
    N = len(good_texts)
    if (l == 1):
        freq = monogram_freq(text)
    elif (l == 2):
        freq = bigram_freq(text)
    for i in range(N):
        if (criterion_2_3(good_texts[i], top, freq, l, h) == 0):
            false_neg += 1
        if (criterion_2_3(bad_texts[i], top, freq, l, h) == 1):
            false_pos += 1
    return false_neg/N, false_pos/N
```

```python
def criterion_4_0(test, theor_index, l = 1, k = 0.001):
    status = 1
    if (l != 1) and (l != 2):
        raise ValueError('L must be 1 or 2.')
    emp_index = index(test, l)
    diff = abs(emp_index - theor_index)
    if (diff > k):
        status = 0
        return status
    return status
```

```python
def forall_4_0(text, good_texts, bad_texts, l = 1, k = 0.001):
    false_pos = 0
    false_neg = 0
    N = len(good_texts)
    theor_index = index(text, l)
    for i in range(N):
        if (criterion_4_0(good_texts[i], theor_index, l, k) == 0):
            false_neg += 1
        if (criterion_4_0(bad_texts[i], theor_index, l, k) == 1):
            false_pos += 1
    return false_neg/N, false_pos/N
```
